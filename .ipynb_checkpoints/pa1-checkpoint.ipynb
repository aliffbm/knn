## Import libraries/data and create dfs

```python
import pandas as pd
import numpy as np 
from sklearn.neighbors import KNeighborsClassifier
import time
import copy
```

```python
# pull in files and create dataframes
ndArray = np.loadtxt("pa1train.txt")
project = np.loadtxt("projection.txt")
validation = np.loadtxt("pa1validate.txt")
test = np.loadtxt("pa1test.txt")

df = pd.DataFrame(ndArray)
validationDF = pd.DataFrame(validation)
testDF = pd.DataFrame(test)
projectDF = pd.DataFrame(project)

X = df.drop([784], axis=1)
y = df[784]

X_validation = validationDF.drop([784], axis=1)
y_validation = validationDF[784]

X_test = testDF.drop([784], axis=1)
y_test = testDF[784]
```

```python
# create projectsion for training, validation, and test
projectedX = X.dot(projectDF)
projectedX_validation = X_validation.dot(projectDF)
projectedX_test = X_test.dot(projectDF)
```

# Problem 1: 
## Which of these classifiers performs the best on validation data?
###### KNN classifier where k=1 performs the best on validation as it has the lowest error rate of 0.082.

## What is the test error of this classifier?
###### The test error of this classifier is .093.


#### Below is code and table output



```python
errorsDF = {"k": [],"training_errors": [], "validation_errors":[]}
bestPerform = None
prevValidPred = None
bestK = None
k = 16
beginTime = time.time()
for k in range(1, 16): 
    neigh = KNeighborsClassifier(n_neighbors=k)
    neigh.fit(X, y)
    
    # predict on training
    pred = neigh.predict(X)
    error = 1 - ((pred == y).sum() / len(pred))
    errorsDF["training_errors"].append(error)
    
    #pred on validation
    pred = neigh.predict(X_validation)
    error = 1 - ((pred == y_validation).sum() / len(pred))
    errorsDF["validation_errors"].append(error)
    
    #add what k neighbors
    errorsDF["k"].append(k)
    
    # get the best classifier and save in bestPerform
    if prevValidPred == None:
        prevValidPred = error
        bestPerform = neigh
        bestK = k
    elif error < prevValidPred:
        prevValidPred = error
        bestPerform = neigh
        bestK = k
    
    
    
    
endTime = time.time()
totalTime = endTime - beginTime
print("Total Time to compute: ", totalTime)
print("Best Classifier in validation is with k = ", bestK, " with a error rate of: ", prevValidPred)
test_pred = bestPerform.predict(X_test)
errorRate = 1 - ((test_pred == y_test).sum() / len(test_pred))
print("The test error on classifier with k=", bestK, " is ", errorRate)

edf = pd.DataFrame(errorsDF)
edf["k"] = edf.k.apply(lambda x: 'k={0}'.format(x))
edf = edf.set_index('k')
edf
```

# Problem 2:
## How is the classification accuracy affected by projection? 
###### The classification error on K, generally, increased, thus the accuracy decreased. 

## How does the running time of your program change when you run it on projected data?
###### It significantly lowers the running time by ~ 108 seconds. 

##### Test error is: .296
###### Below is code and table output for projected data

```python
errorsProjDF = {"k": [],"training_errors": [], "validation_errors":[]}
bestPerform = None
prevValidPred = None
bestK = None
k = 16

beginTime = time.time()
for k in range(1, 16): 
    projNeigh = KNeighborsClassifier(n_neighbors=k)
    projNeigh.fit(projectedX, y)
    pred = projNeigh.predict(projectedX)
    error = 1 - ((pred == y).sum() / len(pred))

    errorsProjDF["training_errors"].append(error)
    
     #pred on validation
    pred = projNeigh.predict(projectedX_validation)
    error = 1 - ((pred == y_validation).sum() / len(pred))
    errorsProjDF["validation_errors"].append(error)
    
    errorsProjDF["k"].append(k)
    
     # get the best classifier and save in bestPerform
    if prevValidPred == None:
        prevValidPred = error
        bestPerform = projNeigh
        bestK = k
    elif error < prevValidPred:
        prevValidPred = error
        bestPerform = projNeigh
        bestK = k
    
    
endTime = time.time()
totalTime = endTime - beginTime
print("Total Time to compute: ", totalTime)
print("Best Classifier in validation is with k = ", bestK, " with a error rate of: ", prevValidPred)
test_pred = bestPerform.predict(projectedX_test)
errorRate = 1 - ((test_pred == y_test).sum() / len(test_pred))
print("The test error on classifier with k=", bestK, " is ", errorRate)

edpf = pd.DataFrame(errorsProjDF)
edpf["k"] = edpf.k.apply(lambda x: 'k={0}'.format(x))
edpf = edpf.set_index('k')
edpf
```
